```python
import csv 

list_URL_WIKI = []
with open("./ville_avec_chevron.csv") as f:   
    for c in f:
        list_URL_WIKI.append(c)
```

```python
print(len(list_URL_WIKI))
```

```python
# ici, je vais chercher les URL correspondant dans DBpedia en utilsant un requete.
# et s'il existe, j'ai enregistré le URL de DB et si non j'ai noté "vide".
import time

list_URL_DB = []

from SPARQLWrapper import SPARQLWrapper, JSON

sparql = SPARQLWrapper("http://dbpedia.org/sparql")

for i in list_URL_WIKI:
    
    time.sleep(3)

    sparql.setQuery("""
    select ?Concept
    where {?Concept owl:sameAs  %s}
"""%(i))
    
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    
    v = results["results"]["bindings"]
    
    if len(v)==0:
        list_URL_DB.append("vide")

    if len(v)!=0:

        d = v[0]["Concept"]["value"]
        
  
        list_URL_DB.append(d)
    
```

```python
# les longueurs pour les 2 listes sont égales 
print(len(list_URL_WIKI))
print(len(list_URL_DB))
print(list_URL_DB[:10])
```

```python
# ici, j'ai enregistré les URL ville WIKI dans la liste list_URL_ville_wiki
import json
with open("ville_URL_Nom_WIKI.json") as f:
    dic = json.load(f)
    
list_nom_ville = []
list_URL_ville_wiki = []
for c,v in dic.items():
    list_nom_ville.append(v)
    list_URL_ville_wiki.append(c)
    
```

```python
import pandas as pd
df = pd.read_csv("~/Desktop/M/DATA/film/film_nom_2_URL.csv")

list_URL_film_DB = []

for i in df["URL de Dbpedia"] :
    list_URL_film_DB.append(i)
 
```

```python
import re
df = pd.read_csv("~/Desktop/M/DATA/film/film_nom_2_URL.csv")

list_nom_film_DB = []

for i in list_URL_film_DB :
    
    c = re.search(r'''(?<=ce/).*?$''',i)
    nom = c.group(0)
    list_nom_film_DB.append(nom)
```

```python
print(list_nom_film_DB)
```

```python
print(list_URL_film_DB)
```

```python
#J'ai récupéré les plongements et les ai enrgistré dans le dictionnaire Nom_emd
import numpy as np

Nom_emd = {}
emd = []
for i in list_nom_film_DB:

    c = "dbr:"+str(i)#j'ai ajouté dbr (préfiex de IRI http://dbpedia.org/resource/) devant le nom de l'entité.
    print(c)
    if c in model.wv.vocab:
        e = model.wv[c]
        emd.append(e.tolist())
    else : 
        emd.append(np.zeros(200).tolist())
        
Nom_emd= dict(zip(list_nom_film_DB,emd))

print(len(Nom_emd))
print(type(Nom_emd))

```

```python
np.save('film_Nom_emd_DB_rdf2vec.npy', Nom_emd)
read_dictionary = np.load('film_Nom_emd_DB_rdf2vec.npy', allow_pickle=True).item()
print(read_dictionary["Anna_Karenina_(2012_film)"])
```

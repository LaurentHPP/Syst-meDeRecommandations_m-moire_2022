```python
import csv 

list_URL_WIKI = []
with open("./cusine_avec_chevron.csv") as f:   
    for c in f:
        list_URL_WIKI.append(c)
```

```python
print(len(list_URL_WIKI))
```

```python
# ici, je vais chercher les URL correspondant dans DBpedia en utilsant un requete.
# et s'il existe, j'ai enregistré le URL de DB et si non j'ai noté "vide".
import time

list_URL_DB = []

from SPARQLWrapper import SPARQLWrapper, JSON

sparql = SPARQLWrapper("http://dbpedia.org/sparql")

for i in list_URL_WIKI:
    
    time.sleep(1)

    sparql.setQuery("""
    select ?Concept
    where {?Concept owl:sameAs  %s}
"""%(i))  
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    print(results)
    v = results["results"]["bindings"]
    if len(v)==0:
        list_URL_DB.append("vide")
        print(v)
    if len(v)!=0:
        print(v)
        d = v[0]["Concept"]["value"]
        list_URL_DB.append(d)
    
```

```python
# les longueurs pour les 2 listes sont égales 
print(len(list_URL_WIKI))
print(len(list_URL_DB))
```

```python
# ici, j'ai enregistré les URL cusine WIKI dans la liste list_URL_cusine_wiki
import json
with open("cusine_URL_Nom_WIKI.json") as f:
    dic = json.load(f)
    
list_nom_cusine = []
list_URL_cusine_wiki = []
for c,v in dic.items():
    list_nom_cusine.append(v)
    list_URL_cusine_wiki.append(c)
    
```

```python
print(list_URL_cusine_wiki)
```

```python
print(len(list_URL_cusine_wiki))
print(len(list_nom_cusine))
print(len(list_URL_DB))
print(len(list_URL_DB))
```

```python
#j'ai créé un fichier csv, il y a 3 colonnes : Nom de cusine, URL de Dbpedia et URL de Wikipedia
import csv 
import pandas as pd
cusine_nom_2_URL = []
for i in range(len(list_nom_cusine)):
    cusine_nom_2_URL.append([list_nom_cusine[i],list_URL_DB[i],list_URL_cusine_wiki[i]])

column_name = ["Nom de cusine","URL de Dbpedia","URL de Wikipedia"]
csv_name = "cusine_nom_2_URL.csv"
xml_df = pd.DataFrame(cusine_nom_2_URL,columns = column_name)
xml_df.to_csv(csv_name,index=None)

    
    
```

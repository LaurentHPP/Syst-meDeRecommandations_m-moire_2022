```python
# J'ai téléchargé les fichiers sur ce lien : http://data.dws.informatik.uni-mannheim.de/rdf2vec/models/DBpedia/2015-10/4depth/skipgram/
# et je les ai mis dans un même répertoire
# j'ai chargé le modèle en utilisant la bibliothèque gensim

import gensim
model = gensim.models.Word2Vec.load("DB2Vec_sg_200_5_5_15_4_500")
```

```python
import pandas as pd
df = pd.read_csv("~/Desktop/M/DATA/cusine/cusine_nom_2_URL.csv")

list_URL_cusine_DB = []

for i in df["URL de Dbpedia"] :
    list_URL_cusine_DB.append(i)
 
```

```python
print(list_URL_cusine_DB[:10])
```

```python
import re
df = pd.read_csv("~/Desktop/M/DATA/cusine/cusine_nom_2_URL.csv")

list_nom_cusine_DB = []

for i in list_URL_cusine_DB :
    #print(i)
    c = re.search(r'''(?<=ce/).*?$''',i)
    nom = c.group(0)
    list_nom_cusine_DB.append(nom)
    
    
     
    
```

```python
print(list_nom_cusine_DB)
```

```python
#J'ai récupéré les plongements et les ai enrgistré dans le dictionnaire Nom_emd
import numpy as np

Nom_emd = {}
emd = []
for i in list_nom_cusine_DB:

    c = "dbr:"+str(i)#j'ai ajouté dbr (préfiex de IRI http://dbpedia.org/resource/) devant le nom de l'entité.
    print(c)
    if c in model.wv.vocab:
        e = model.wv[c]
        emd.append(e.tolist())
    else : 
        emd.append(np.zeros(200).tolist())
        
Nom_emd= dict(zip(list_nom_cusine_DB,emd))

print(len(Nom_emd))
print(type(Nom_emd))
#print(Nom_emd)
```

```python
np.save('cusine_Nom_emd_DB_rdf2vec.npy', Nom_emd)
read_dictionary = np.load('cusine_Nom_emd_DB_rdf2vec.npy', allow_pickle=True).item()
print(read_dictionary["Sushi"])
```

```python
# in : query_cusine_wiki.json
# out: NOM_URL.json
#      URL_NOM.json  
#      cusine_avec_chevron.csv   
 

# ce script est pour convertir le fichier json que j'ai téléchargé via wikidata 
# en un dictionnaire dont le key est URL du films et la valeur est son nom.
# en un dictionnaire dont le key est nom du films et la valeur est son URL.
# et en le fichier film_avec_chevron.csv dans lequel les URL des entités ont tous les chevrons.
```

```python
import json
```

```python
# le fichier query.json est ce que j'ai téléchargé sur Web Wikidata Query Service
chemin = "query.json"
```

```python
labels = open(chemin, "r", encoding="UTF-8")
```

```python
label_nom = json.load(labels )
```

```python
print(len(label_nom))
print(label_nom[:10])
```

```python
dic_label_nom = {}
for i in label_nom:
    #print(i["item"])
    #print(i["itemLabel"])
    dic_label_nom["<"+i["item"]+">"] = [i["itemLabel"]]
```

```python
print(len(dic_label_nom))
```

```python
# J'enregistre les données dans le fichier cusine_URL_Nom_WIKI.json
a = json.dumps(dic_label_nom, ensure_ascii=False)
f = open("cusine_URL_Nom_WIKI.json","w", encoding="UTF-8")
f.write(a)
f.close()
```

```python
dic_nom_label = {}
for i in label_nom:
    #print(i["item"])
    #print(i["itemLabel"])
    dic_nom_label[i["itemLabel"]] = "<"+i["item"]+">"# j'ajoute les chevrons pour les URL de wikipedia
```

```python
print(dic_nom_label)
```

```python
# j'enregistre les données dans le fichier cusine_Nom_URL_WIKI.json
a = json.dumps(dic_nom_label, ensure_ascii=False)
f = open("cusine_Nom_URL_WIKI.json","w", encoding="UTF-8")
f.write(a)
f.close()
```

```python
import gensim
from gensim.test.utils import datapath, get_tmpfile
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec
from gensim.models import Word2Vec 
import os
import json
import re
```

```python
# Convert glove format to word2vec
gensim.scripts.glove2word2vec.glove2word2vec('100_films_word2vec.txt', 'test_word2vec.txt')
```

```python
# dans les plongements que j'ai récupérés via wikidata, il y a plusieurs espaces entre certaine dimentions, 
#cela m'empeche de importer les plogements dans le model. donc je vais nettoyer les espaces 
path = "test_word2vec.txt"
def clean_fichier(path):
    clean_chaine_txt = " "
    f = open(path)
    chaines = f.readlines()
    for chaine in chaines:
        clean_chaine = " ".join(chaine.split())
        clean_chaine_txt+=clean_chaine +"\n" 

        #clean_chaine_txt  + "\n" + clean_chaine
    f.close()
    return clean_chaine_txt
```

```python
clean_chaine_txt = clean_fichier(path)
```

```python
# cette étaper est pour le but de sauvegarder un bon format du model
with open("100_films_word2vec_clean.txt","w",encoding = 'utf-8') as f:
    f.write(clean_chaine_txt)   
```

```python
path = "100_films_word2vec_clean.txt"
```

```python
model= KeyedVectors.load_word2vec_format(path)
```

```python
#importer dictionnaire label_nom(key -> URL du films , value -> nom du film)
label_nom = open("100_film_URL_Nom.json",encoding = "utf-8")
dic_label_nom = json.load(label_nom)
```

```python
#importer dictionnaire nom_label(key -> nom du film , value -> URL du films)
label_nom = open("100_film_Nom_URL.json",encoding = "utf-8")
dic_nom_label = json.load(label_nom)
```

```python
def recommander():
    print("Quel est votre film préféré ?:" )
    nom_pos_input = input()
    nom_pos = dic_nom_label[nom_pos_input]
    print("Quel est le film que vous n'aimez pas ?:" )
    nom_neg_input = input()
    nom_neg = dic_nom_label[nom_neg_input]
    sims = model.most_similar(positive = nom_pos,negative = nom_neg, topn = 10)
    #print(sims)
    list_recom = []
    print("="*50)
    print("Je vous conseille de voir les films suivants :")
    for i in range(len(sims)):
        if sims[i][0] in dic_label_nom:
            list_recom.append(dic_label_nom[sims[i][0]])
            print(dic_label_nom[sims[i][0]])
    return list_recom
    
```

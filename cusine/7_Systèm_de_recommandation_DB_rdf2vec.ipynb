```python
import pandas as pd
df = pd.read_csv("cusine_nom_2_URL.csv")
```

```python
dic = {}
for i in range(len(df["Nom de cusine"])):
    c = df["Nom de cusine"][i]
    v = df["URL de Dbpedia"][i]
    dic[c] = v
    
dic_1 = {}
for c, v in dic.items():
    cle = c[2:-2].replace(" ","_")
    cle = cle[0].upper() + cle[1:]
    dic_1[cle] = v
```

```python
print(dic_1["Tong_Sui_Shop"])
```

```python
# la fonction qui nous permet de générer la recommandation 

def recommander(pos,neg):
    sims = model.most_similar(positive = pos,negative =neg, topn = 5)
    print("Je vous conseille de manger les cusines suivantes :")
    print("\n")
    print("="*50)
    print("\n")
    for i in range(len(sims)):
        print("Numéro",i+1,":",sims[i][0])
        print("Pour plus de l'information, cliquez : ", dic_1[sims[i][0]])
        print("\n")
    
```

```python
recommander(["Baozi","Matzah_pizza"],["Doner_kebab"])
```

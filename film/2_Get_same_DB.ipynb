```python
import csv 

list_URL_WIKI = []
with open("./film_avec_chevron.csv") as f:   
    for c in f:
        list_URL_WIKI.append(c)

   
```

```python
print(len(list_URL_WIKI))
```

```python
# ici, je vais chercher les URL correspondant dans DBpedia en utilsant un requete.
# et s'il existe, j'ai enregistré le URL de DB et si non j'ai noté "vide".
import time
list_URL_DB = []
from SPARQLWrapper import SPARQLWrapper, JSON
sparql = SPARQLWrapper("http://dbpedia.org/sparql")
for i in list_URL_WIKI:
    time.sleep(1)
    sparql.setQuery("""
    select ?Concept
    where {?Concept owl:sameAs  %s}
"""%(i))
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    print(results)
    v = results["results"]["bindings"]
    if len(v)==0:
        list_URL_DB.append("vide")
        print(v)
    if len(v)!=0:
        print(v)
        d = v[0]["Concept"]["value"]
        list_URL_DB.append(d)
```

```python
# ici, j'ai enregistré les URL film WIKI dans la liste film_URL_Nom_WIKI
import json
with open("film_URL_Nom_WIKI.json") as f:
    dic = json.load(f)
    
list_nom_film = []
list_URL_film_wiki = []
for c,v in dic.items():
    list_nom_film.append(v)
    list_URL_film_wiki.append(c)
    
```

```python
#j'ai créé un fichier csv, il y a 3 colonnes : Nom de film, URL de Dbpedia et URL de Wikipedia

import csv 
import pandas as pd
film_nom_2_URL = []
for i in range(len(list_nom_film)):
    film_nom_2_URL.append([list_nom_film[i],list_URL_DB[i],list_URL_film_wiki[i]])

column_name = ["Nom de film","URL de Dbpedia","URL de Wikipedia"]
csv_name = "film_nom_2_URL.csv"
xml_df = pd.DataFrame(film_nom_2_URL,columns = column_name)
xml_df.to_csv(csv_name,index=None)

    
    
```

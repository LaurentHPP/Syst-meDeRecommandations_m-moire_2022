```python
import pandas as pd
import re
```

```python
df = pd.read_csv("~/Desktop/M/DATA/film/film_nom_2_URL.csv")

list_URL_film_DB = []

for i in df["URL de Dbpedia"] :
    list_URL_film_DB.append(i)
    

list_nom_film_DB = []

for i in list_URL_film_DB :
    
    c = re.search(r'''(?<=ce/).*?$''',i)
    nom = c.group(0)
    list_nom_film_DB.append(nom)
    
list_nom_film = []
for i in df["Nom de film"] :
    list_nom_film .append(i[2:-2].replace(" ","_"))
```

```python
print(len(list_nom_film_DB))
print(len(list_nom_film))
dict_nom_nomDB = dict(zip(list_nom_film,list_nom_film_DB))
dict_nomDB_nom = dict(zip(list_nom_film_DB,list_nom_film))
```

```python
print(dict_nom_nomDB["L'Homme_qui_a_tué_Don_Quichotte"])
```

```python



dic = {}
for i in range(len(df["Nom de film"])):
    c = df["Nom de film"][i]
    v = df["URL de Dbpedia"][i]
    dic[c] = v
    
dic_1 = {}
for c, v in dic.items():
    cle = c[2:-2].replace(" ","_")
    cle = cle[0].upper() + cle[1:]
    dic_1[cle] = v
```

```python
print(dic_1["Vie_sauvage"])
```

```python
# la fonction qui nous permet de générer la recommandation 

def recommander(pos,neg):
    pos_1 = []
    if len(pos) != 0:
        for i in pos:
            pos_1.append(dict_nom_nomDB[i])
    neg_1 = []
    if len(neg) != 0:
        for i in neg:
            neg_1.append(dict_nom_nomDB[i])
            
    sims = model.most_similar(positive = pos_1,negative =neg_1, topn = 5)
    print("Je vous conseille de voir les films suivants :")
    print("\n")
    print("="*50)
    print("\n")
    for i in range(len(sims)):
        print("Numéro",i+1,":",dict_nomDB_nom[sims[i][0]])
        print("Pour plus de l'information, cliquez : ", dic_1[dict_nomDB_nom[sims[i][0]]])
        print("\n")
    
```

```python
recommander(["Anna_Karénine","La_Religieuse"],["Twentynine_Palms"])
```
